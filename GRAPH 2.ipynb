# Kruskal's Algorithm

```python
class edge:
    def  __init__(self,src,dest,wt):
        self.src = src
        self.dest = dest
        self.wt = wt
        
def getParent(v,parent):
    
    if (v == parent[v]):
        return v
    return getParent(parent[v],parent)

def kruskal(edges,nVertices):
    parent = [i for i in range (nVertices)]
    edges = sorted(edges,key=lambda edge:edge.wt)
    
    count = 0
    output = []
    i = 0
    while count < (nVertices - 1):
        currentEdge = edges[i]
        srcParent = getParent(currentEdge.src,parent)
        destParent = getParent(currentEdge.dest,parent)
        
        if scrParent != destParent:
            output.append(currentEdge)
            count+=1
            parent[srcParent]=destParent
            
        i += 1
        
    return output
```

# Dijkstra's Algorithm

```python
def getVertex():
    min = 2*(10)**31
    for i in range(v):
        if visited[i] == False and dist[i] < min :
            min = dist[i]
            index = i
            
    return index

def dijkstra():
    for i in range (v-1):
        vertex = getVertex()
        visited[vertex]= True
        for j in range(v):
            if matrix[vertex][j] > 0 and visited[j] == False:
                if dist[vertex]+matrix[vertex][j] < dist[j]:
                    dist[j] = dist[vertex] + matrix[vertex][j]
                    
                    
v,e = [int(i) for i in input().split(' ')]
matrix = [[0 for i in range(v)] for j in range (v)]
visited = [False for i in range (v)]
dist = [(2*(10)**31)for i in range(v)]
dist[0] = 0
sv = 0
for i in range(e):
    v1,v2,dis = [int(i) for i in input().split(' ')]
    matrix[v1][v2]= dis
    matrix[v2][v1]= dis
    
dijkstra()

for i in range(v):
    print(i,dist[i])
```
